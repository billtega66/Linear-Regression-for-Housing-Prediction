```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

Load Data

```python
data = pd.read_csv("housing[1].csv")
#data.head()
```

```python
unique_sorted_values = sorted(data['ocean_proximity'].unique())  # Replace 'your_column_name' with your column
print(unique_sorted_values)
```

Change Ocean Proximity

```python
#data = data.drop(columns=['ocean_proximity'])
data['ocean_proximity'] = data['ocean_proximity'].replace({'NEAR BAY': 1, 'INLAND': 0, '<1H OCEAN': 2, 'NEAR OCEAN': 3, 'ISLAND': 4})

```

```python
data
```

```python
data['ocean_proximity'] = data['ocean_proximity'].astype(float)
```

Drop NaN Values

```python
data = data.dropna()
```

```python
print(data.dtypes)

```

Define Target Variable

```python
X = data.drop(columns=['median_house_value'])
y = data['median_house_value']
```

Split Data Into Training and Testing

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

Scaling the Data

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaler = scaler.fit_transform(X_train)
X_test_scaler = scaler.transform(X_test)
```

Model Development and Evaluation

```python
from sklearn.linear_model import LinearRegression
linear_reg = LinearRegression()
```

Polynomial Regression Degree 2

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=2)
X_poly_train = poly_reg.fit_transform(X_train_scaler)
X_test_poly = poly_reg.transform(X_test_scaler)
poly_reg.fit(X_poly_train,y_train)
linear_reg.fit(X_poly_train,y_train)
```

Make Prediction

```python
y_pred = linear_reg.predict(X_test_poly)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test,y_pred)
```

```python
y_pred_train = linear_reg.predict(X_poly_train)
mean_absolute_error(y_train,y_pred_train)
```

Mean Squared Error / R^2 Score

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'R^2 Score: {r2}')
```

Polynomial Regression degree 3 Results

```python
poly_reg = PolynomialFeatures(degree=3)
X_poly_train = poly_reg.fit_transform(X_train_scaler)
X_test_poly = poly_reg.transform(X_test_scaler)
poly_reg.fit(X_poly_train,y_train)
linear_reg.fit(X_poly_train,y_train)
```

```python
y_pred =  linear_reg.predict(X_test_poly)
mean_absolute_error(y_test, y_pred)
```

```python
y_pred_train =  linear_reg.predict(X_poly_train)
mean_absolute_error(y_train, y_pred_train)
```

```python
poly_reg = PolynomialFeatures(degree=4)
X_poly_train = poly_reg.fit_transform(X_train_scaler)
X_test_poly = poly_reg.transform(X_test_scaler)
poly_reg.fit(X_poly_train,y_train)
linear_reg.fit(X_poly_train,y_train)
```
